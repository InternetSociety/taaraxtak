```python
from typing import Tuple
def extract_from_fn (csv_path) -> Tuple[pd.Timestamp, str]:
    '''
    Return tuple of 
    - time of measurement (timestamp)
    - name of market (string)
    
    NOTE - This is the edge, where validation should happen.
    '''
    csv_fn = csv_path.split('/')[-1].split('.csv')[0]
    date, market = csv_fn.split('__')
    return (pd.to_datetime(date), market)

ex_fn = 'csv-data/daily-data/2021-01-05__data-centers.csv'
time, market = extract_from_fn(ex_fn) 
```

```python
from datetime import timedelta
def offset (t: pd.Timestamp) -> pd.Timestamp:
    return t + timedelta(hours=9)
```

# Daily data

```python
from typing import Optional

ex_df = pd.read_csv(ex_fn)
utils.extract_from_row(market, time, ex_df.iloc[0])
```

```python
from os import listdir
from os.path import join
from funcy import partial

my_dir = join('csv-data', 'daily-data')
for file in listdir(my_dir):
    print(file)
    file_fn = join(my_dir, file)
    time, market = extract_from_fn(file_fn) 
    # HACK - offset time to roughly align it with when our data were collected.
    time = offset(time)
    extract = partial(utils.extract_from_row, market, time)
    df = pd.read_csv(file_fn)
    marketshares = df.apply(extract, axis=1)
    [ marketshare.write_to_db(cur, conn, commit=False)
    for marketshare in marketshares ]
conn.commit()
```

# Monthly data

```python
import numpy as np
from typing import Generator

def extract_from_monthly_row (market, row: pd.Series) -> Generator[types.ProviderMarketshare, None, None]:
    url = None
    name = row[0]
    jurisdiction_alpha2 =  utils.get_country(name)
    for time, marketshare in row[1:].items():
        # validate time
        timestamp = pd.to_datetime(time)
        # convert marketshares to be 0-1
        marketshare = marketshare/100
        pm = types.ProviderMarketshare(name, None, jurisdiction_alpha2, market, float(marketshare), timestamp)
        pm.write_to_db(cur, conn, commit=False)
        
ex_fn_monthly = 'csv-data/monthly-data/2021-02-01__ssl-certificate.csv'
ex_df = pd.read_csv(ex_fn_monthly).fillna(0)
ex_row = ex_df.iloc[5]
extract_from_monthly_row('ssl-certificate', ex_row)
```

```python
from funcy import flatten

my_dir = join('csv-data', 'monthly-data')
for file in listdir(my_dir):
    print(file)
    file_fn = join(my_dir, file)
    _, market = extract_from_fn(file_fn) 
    df = pd.read_csv(file_fn).fillna(0)
    extract = partial(extract_from_monthly_row, market)
    df.apply(extract, axis=1)
conn.commit()
```
